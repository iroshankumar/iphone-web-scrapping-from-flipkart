```python
#importing libraryes

import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
import logging
```

```python
#creating url for website
flipkarturl="https://www.flipkart.com/search?q="+"iphone+14"
```

```python
flipkarturl
```

```python
#url request
urlclint=uReq(flipkarturl)
```

```python
#reading page
flipkart_page=urlclint.read()
```

```python
#butifying website page using BeautifulSoup
flipkart_html=bs(flipkart_page,'html.parser')
```

```python
flipkart_html
```

```python
#products scraping

bigbox=flipkart_html.find_all('div',{'class':"_1AtVbE col-12-12"})
```

```python
len(bigbox)
```

```python
product_link="https://www.flipkart.com"+bigbox[1].div.div.div.a['href']
```

```python
product_request=requests.get(product_link)
```

```python
product_html=bs(product_request.text,'html.parser')
```

```python
#reading name of the person who comment

comment=product_html.find_all('div',{"class":"_16PBlm"})
```

```python
comment[0].div.div.find_all('p',{"class":"_2sc7ZR _2V5EHH"})[0].text
```

```python
for i in comment:
    print(i.div.div.find_all('p',{"class":"_2sc7ZR _2V5EHH"})[0].text)
```
